# **Checking and Handling Null values**

```python
titanic_data.isnull().sum()
```

```python
#let's handle null value in age column
titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].mean())
```

```python
#lets handle null value in Cabin
# filling the null values with the value of next non-null value in column
titanic_data['Cabin'] = titanic_data['Cabin'].ffill()
```

```python
# let's handle the null values in Embarked column
titanic_data['Embarked'] = titanic_data['Embarked'].bfill()
```

```python
titanic_data.dropna(inplace=True)
```
